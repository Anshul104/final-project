```python
import pandas as pd
import geopandas
%matplotlib inline
```

```python

# import dataset
grounds = pd.read_csv("PlayGroundData.csv", low_memory=False)
# filter to important info
num_grounds = grounds.groupby("neighborhood").sum()
num_grounds.head()
```

```python

# do the merge
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods.shp")
ground_map = neighborhoods.merge(num_grounds, how='left', left_on='hood', right_on='neighborhood')

#graph
ground_map.plot(column='grounds', # set the data to be used for coloring
                cmap = 'Purples', # choose a color palette
               edgecolor="black",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Number of Playgrounds"}, # label the legend
               figsize=(15, 10),# set the size
               ) 
```
